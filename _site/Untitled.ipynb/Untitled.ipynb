```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
surveys_complete = pd.read_csv('/Users/argyris/Downloads/surveys.csv')
surveys_complete = surveys_complete.dropna()
```

```python
plt.scatter(x=surveys_complete["weight"], y=surveys_complete["hindfoot_length"])
plt.savefig("../fig/matplotlib_scatter.png")
```

```python

sns.scatterplot(data=surveys_complete, x="weight", y="hindfoot_length")
plt.savefig("../fig/seaborn_scatter.png")
```

```python
sns.scatterplot(data=surveys_complete, x="weight", y="hindfoot_length", alpha=0.25)
plt.savefig("../fig/seaborn_scatter_alpha.png")
```

```python
sns.scatterplot(data=surveys_complete, x="weight", y="hindfoot_length", alpha=0.25, color="red")
plt.savefig("../fig/seaborn_scatter_alpha_red.png")
```

```python
sns.scatterplot(data=surveys_complete, x="weight", y="hindfoot_length", hue="species_id", alpha=0.5)
plt.legend(ncol=3)
plt.savefig("../fig/seaborn_scatter_alpha_speciesid.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot()
sns.scatterplot(data=surveys_complete, x="weight", y="hindfoot_length",
                alpha=0.5, hue="species_id", ax=ax)
ax.set_xlabel("Weight (g)")
plt.legend(ncol=3)
plt.savefig("../fig/seaborn_scatter_alpha_speciesid_xlabel.png")
```

```python
sns.boxplot(data=surveys_complete, x="species_id", y="weight")
plt.savefig("../fig/seaborn_dist.png")
```

```python
sns.stripplot(data=surveys_complete, x="species_id", y="weight")
plt.savefig("../fig/seaborn_dist_strip.png")
```

```python
yearly_counts = surveys_complete.groupby(['year', 'species_id'])['species_id'].count()
yearly_counts = yearly_counts.reset_index(name='counts')
sns.lineplot(data=yearly_counts, x='year', y='counts', hue='species_id')
plt.legend(ncol=3)
plt.savefig("../fig/seaborn_lineplots.png")
```

```python
sns.scatterplot(data=surveys_complete, x="weight", y="hindfoot_length")
```

```python
sns.relplot(data=surveys_complete, x="weight", y="hindfoot_length",
            col="sex", hue="species_id")
plt.savefig("../fig/seaborn_facet1.png")
```

```python
sns.relplot(data=surveys_complete, x="weight", y="hindfoot_length",
            col="plot_id", hue="species_id", col_wrap=5)
plt.savefig("../fig/seaborn_facet2.png")
```

```python
sns.countplot(data=surveys_complete, x="year")
plt.savefig("../fig/countplot.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot()
sns.countplot(data=surveys_complete, x="year", ax=ax)
ax.tick_params(axis='x', rotation=90)
plt.savefig("../fig/countplot_rot.png")
```
